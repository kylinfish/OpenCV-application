## Translation

You can take make it into a Numpy array of type np.float32 and pass it into cv2.warpAffine() function. <br>See below example for a shift of <b>(150,80):

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
import cv2

img = cv2.imread('123.jpg')

#if image is color, it will return [rows], [columns] and [channels] 
rows,cols,channels = img.shape
# here is Matrix coordinate , translate 150 ,80
M = np.float32([[1,0,150],[0,1,80]]) 
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

## Affine Transformation

Then <b>cv2.getAffineTransform</b> will create a <b>2x3 matrix</b> which is to be passed to <b>cv2.warpAffine.</b>

```python
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

#matrix
M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

## Perspective Transformation (透視）

Then transformation matrix can be found by the function <b>cv2.getPerspectiveTransform</b>.<br>
Then apply <b>cv2.warpPerspective</b> with this 3x3 transformation matrix.

```python
pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(250,250))
# return 3*3 matrix
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

## Canny Edge Detection

OpenCV functions for that : cv2.Canny()
<ol>
<li>Noise reduction</li>
<li>Finding Intensity Gradient of the Image</li>
<li>Non-maximum Suppression</li>
<li>Hysteresis Thresholding</li></ol>

```python
img = cv2.imread('123.jpg',0)
edges = cv2.Canny(img,100,150)
edges2 = cv2.Canny(img,100,120)
#subplot# 1x3 grid, indexing plot

plt.subplot(131),plt.imshow(img,cmap = 'gray')
plt.title('Original '), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(edges,cmap = 'gray')
plt.title('Edge: 100, 150'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(edges2,cmap = 'gray')
plt.title('Edge: 50, 80'), plt.xticks([]), plt.yticks([])

#spacing 
plt.subplots_adjust(left=0.1, right=0.9,
                    bottom=0.1, top=0.9,
                    hspace=0.4, wspace=0.4)
plt.show()
```
