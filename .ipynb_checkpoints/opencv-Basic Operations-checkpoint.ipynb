## Accessing and Modifying pixel values

<p style="color:blue"> For BGR image, it returns an array of Blue, Green, Red values.</p>

```python
import cv2
import numpy as np

img = cv2.imread('123.jpg')
px = img[100,100]
# print px
print "B:",px[0]," G:",px[1]," R:",px[2]
```

<p style="color:blue">accessing only blue pixel</p>

```python
blue = img[100,100,0]
print blue
```

<p style="color:blue">modify pixel with the same way.</p>

```python
img[100,100] = [255,255,255]
print img[100,100]
```

<p style="color:blue">image shape：It returns a tuple of number of rows, columns and channels (if image is color): </p>

```python
print img.shape
```

## Splitting and Merging Image Channels

```python
b,g,r = cv2.split(img)
img = cv2.merge((b,g,r))
print b, g ,r 
```

```python
#or 
b = img[:,:,0]
print b
```

## Load and display image 

load color image or graylevel image, and display it with windows.<br>


```python
%matplotlib inline
import matplotlib
import numpy as np
import cv2
import sys
from matplotlib import pyplot as plt


image = cv2.imread('123.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # convert image to gray

plt.subplot(121),plt.imshow(image)
plt.subplot(122),plt.imshow(gray_image)
plt.show()
```

## SIFT in OpenCV

```python
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT()
kp = sift.detect(gray,None)

img_sift=cv2.drawKeypoints(gray,kp)

plt.imshow(img_sift)
plt.show()
```

##  cv2.fastNlMeansDenoisingColored()

```python
dst = cv2.fastNlMeansDenoisingColored(img,None,10,10,7,21)

plt.imshow(dst)
plt.show()
```

## SURF 

https://opencv-python-tutroals.readthedocs.org/en/latest/py_tutorials/py_feature2d/py_surf_intro/py_surf_intro.html#surf

```python
img = cv2.imread('123.jpg',0)

# Create SURF object. You can specify params here or later.
# Here I set Hessian Threshold to 400
surf = cv2.SURF(400)

# Find keypoints and descriptors directly
kp, des = surf.detectAndCompute(img,None)


# Check present Hessian threshold
print surf.hessianThreshold


# We set it to some 50000. Remember, it is just for representing in picture.
# In actual cases, it is better to have a value 300-500
surf.hessianThreshold = 50000

# Again compute keypoints and check its number.
kp, des = surf.detectAndCompute(img,None)

print "kp length:",len(kp)
# Recompute the feature points and draw it
kp = surf.detect(img,None)
img2 = cv2.drawKeypoints(img,kp,None,(255,0,0),4)

plt.imshow(img2),plt.show()
plt.show()
```
