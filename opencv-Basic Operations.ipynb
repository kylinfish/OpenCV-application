## Accessing and Modifying pixel values

<p style="color:blue"> For BGR image, it returns an array of Blue, Green, Red values.</p>

```python
import cv2
import numpy as np

img = cv2.imread('123.jpg')
px = img[100,100]
# print px
print "B:",px[0]," G:",px[1]," R:",px[2]
```

```python
img[100,100] = [255,255,255]
print img[100,100]
```

## Bind 2 image to one

```python
import cv2
import numpy as np


alpha = float(input())                 # Ask the value of alpha

if 0<=alpha<=1:                        # Check if 0<= alpha <=1
    beta = 1.0 - alpha                 # Calculate beta = 1 - alpha
    gamma = 0.0                        # parameter gamma = 0

    img1 = cv2.imread('123.jpg')
    img2 = cv2.imread('abc.jpg')

    if img1==None:
        print "img1 not ready"
    elif img2==None:
        print "img2 not ready"
    else:
        dst = cv2.addWeighted(img1,alpha,img2,beta,gamma)  # Get weighted sum of img1 and img2
        #dst = np.uint8(alpha*(img1)+beta*(img2))    # This is simple numpy version of above line. But cv2 function is around 2x faster
        cv2.imshow('dst',dst)
        print "123"
        cv2.waitKey(0)
        cv2.destroyAllWindows()
else:
    print "value of alpha should be 0 and 1"
```

## Change contrast and brightness

```python

import cv2
import numpy as np


alpha = float(input('* Enter the alpha value [1.0-3.0]: '))     # Simple contrast control
beta = int(input('Enter the beta value [0-100]: '))             # Simple brightness control

img = cv2.imread('big.jpg')
mul_img = cv2.multiply(img,np.array([alpha]))                    # mul_img = img*alpha
new_img = cv2.add(mul_img,np.array([beta]))                      # new_img = img*alpha + beta

cv2.imshow('original_image', img)
cv2.imshow('new_image',new_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## SIFT in OpenCV

```python
import cv2
import numpy as np

img = cv2.imread('123.jpg')
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT()
kp = sift.detect(gray,None)

img=cv2.drawKeypoints(gray,kp)
cv2.imshow("",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
#cv2.imwrite('sift_keypoints.jpg',img)
```

## SURF 

https://opencv-python-tutroals.readthedocs.org/en/latest/py_tutorials/py_feature2d/py_surf_intro/py_surf_intro.html#surf

```python
img = cv2.imread('123.jpg',0)

# Create SURF object. You can specify params here or later.
# Here I set Hessian Threshold to 400
surf = cv2.SURF(400)

# Find keypoints and descriptors directly
kp, des = surf.detectAndCompute(img,None)

len(kp)

# Check present Hessian threshold
print surf.hessianThreshold


# We set it to some 50000. Remember, it is just for representing in picture.
# In actual cases, it is better to have a value 300-500
surf.hessianThreshold = 50000

# Again compute keypoints and check its number.
kp, des = surf.detectAndCompute(img,None)

print len(kp)
# Recompute the feature points and draw it
kp = surf.detect(img,None)
img2 = cv2.drawKeypoints(img,kp,None,(255,0,0),4)

plt.imshow(img2),plt.show()
```
